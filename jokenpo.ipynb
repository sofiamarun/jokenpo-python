```python
import random

print("Boas vindas ao jogo de pedra, papel ou tesoura!!")
print("Regras do jogo \nPedra vence de tesoura \nTesoura vence de papel \nPapel vence de pedra")
print("\n")

while True:
  rodadas = input("\n\n Digite quantas rodadas você deseja jogar: ")

  if rodadas.isdigit():
    break
  else:
    print("Você digitou um número inválido para a quantidade de rodadas \nTente novamente e insira um número inteiro")

placar_jogadora = 0
placar_computador = 0

rodadas_int = int(rodadas)
for i in range(1, rodadas_int + 1):
  print(f'\n\nInício da rodada {i}')
  jogada1 = input(f'Jogadora 1, faça a jogada {i}: ').lower()

  jogada1_valida = jogada1 == 'pedra' or jogada1 == 'papel' or jogada1 == 'tesoura'

  while not jogada1_valida:
    print(f'Jogada válida: {jogada1_valida}')
    jogada1 = input(f'Jogadora 1, faça a jogada {i}: ').lower()

    jogada1_valida = jogada1 == 'pedra' or jogada1 == 'papel' or jogada1 == 'tesoura'

  print(f'\nJogadas da rodada {i}: ')
  print(f'- Jogada humana {i}: {jogada1}')

  if jogada1_valida:
    jogadas_validas = ['pedra', 'papel', 'tesoura']
    jogada2 = random.choice(jogadas_validas)
    print(f'- Jogada computador {i}: {jogada2}')

    if jogada1 == jogada2:
      print(f'\nResultado {i}: Empate!')

    elif jogada1 == 'pedra' and jogada2 == 'tesoura':
      print(f'\nResultado {i}: Jogadora 1 ganhou!!')
      placar_jogadora = placar_jogadora + 1

    elif jogada1 == 'papel' and jogada2 == 'pedra':
      print(f'\nResultado {i}: Jogadora 1 ganhou!!')
      placar_jogadora = placar_jogadora + 1

    elif jogada1 == 'tesoura' and jogada2 == 'papel':
      print(f'\nResultado {i}: Jogadora 1 ganhou!!')
      placar_jogadora = placar_jogadora + 1

    else:
      print(f'\nResultado {i}: Computador gamhou!!')
      placar_computador = placar_computador + 1
  else:
    print(f'\nJogada da jogadora 1 foi inválida na rodada {i}')

print("\n\n--> Resultado final: ")
print(f'- Placar Jogadora: {placar_jogadora}')
print(f'- Placar Computador: {placar_computador}')

if placar_jogadora == placar_computador:
  print("\nTemos um empate!")
elif placar_jogadora > placar_computador:
  print("\nParabéns, você venceu!!")
else:
  print("\nQue pena, o ocomputador venceu dessa vez...")

print("Fim de jogo...")
```
